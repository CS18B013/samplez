```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#Boston house prices dataset

from sklearn.datasets import load_boston

X, y = load_boston(return_X_y=True)

#create linear regression model
model = linear_model.LinearRegression()

#splitting data and targets into 70% training and 30% testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,
                                                   random_state=0)

#training the model using training sets
model.fit(X_train, y_train)

#making predictions using testing set
y_pred = model.predict(X_test)

plt.plot(X,y,'b.')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

plt.plot(y_test,y_pred,'c.')
plt.xlabel('Y test')
plt.ylabel('Y predicted')
plt.show()

error = np.mean((y_pred-y_test)**2)
print("ERROR : ",error)
```

```python
from sklearn.datasets import make_moons
from sklearn.linear_model import SGDClassifier

X, y = make_moons()

logistic = SGDClassifier(loss='log')
modify_huber = SGDClassifier(loss = 'modified_huber')
sq_hinge = SGDClassifier(loss = 'squared_hinge')
percep = SGDClassifier(loss = 'perceptron')
sq_loss = SGDClassifier(loss = 'squared_loss')
epsi_insensi = SGDClassifier(loss = 'epsilon_insensitive')
sq_epsi_insensi = SGDClassifier(loss = 'squared_epsilon_insensitive')

logistic.fit(X,y)
modify_huber.fit(X,y)
sq_hinge.fit(X,y)
percep.fit(X,y)
sq_loss.fit(X,y)
epsi_insensi.fit(X,y)
sq_epsi_insensi.fit(X,y)

logistic.score(X,y)
modify_huber.score(X,y)
sq_hinge.score(X,y)
percep.score(X,y)
sq_loss.score(X,y)
epsi_insensi.score(X,y)
sq_epsi_insensi.score(X,y)
```

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_moons

X, y = make_moons()

clf = SGDClassifier(loss="hinge")

clf.fit(X,y)

y_pred = clf.predict(X)

precision, recall, f_score, support = precision_recall_fscore_support(y,y_pred)
tn, fp, fn, tp = confusion_matrix(y,y_pred).ravel()
accuracy = (tp + fn)/(tn+fp+fn+tp)
print("Precision: ",precision)
print("Recall: ",recall)
print("F Score: ",f_score)
print("accuracy: %.5f"%(accuracy))
```
